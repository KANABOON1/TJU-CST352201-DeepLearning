### numpy 的array操作

#### 2.建立一个一维数组 a 初始化为[4,5,6], (1)输出a 的类型（type）(2)输出a的各维度的大小（shape）(3)输出 a的第一个元素（值为4）

```python
# 使用 np.array 函数创建 np.ndarray 类型的对象: 一维数组
a = np.array([4, 5, 6])
print(type(a))
print(a[0])
```

#### 3.建立一个二维数组 b,初始化为 [ [4, 5, 6],[1, 2, 3]] (1)输出各维度的大小（shape）(2)输出 b(0,0)，b(0,1),b(1,1) 这三个元素（对应值分别为4,5,2）

```python
# 使用 np.array 函数创建 np.ndarray 类型的对象: 二维数组
b = np.array([[4, 5, 6], [1, 2, 4]])

print(b.shape)
print(b[0, 0])
print(b[0, 1])
print(b[1, 1])
```

#### 4.  (1)建立一个全0矩阵 a, 大小为 3x3; 类型为整型（提示: dtype = int）(2)建立一个全1矩阵b,大小为4x5;  (3)建立一个单位矩阵c ,大小为4x4; (4)生成一个随机数矩阵d,大小为 3x2.

```python
# 使用 np.zeros 创建形状为 `shape` 的全 0 矩阵, 并使用 dtype=int 约束类型
a = np.zeros((3, 3), dtype=int)
print(a)

# 使用 np.ones 创建形状为 `shape` 的全 1 矩阵
b = np.ones((4, 5))
print(b)


# 使用 np.eye 创建一个 row=N, col=M 的单位矩阵
c = np.eye(N=4, M=4)
print(c)

# 使用 np.random.rand 创建一个形状为 `shape` 的随机矩阵
d = np.random.rand(3, 2)
print(d)
```

#### 24,画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
import matplotlib.pyplot as plt

x = np.arange(0, 100, 0.1)
y = x * x

plt.figure(figsize=(8,5))       # 设置图像大小
plt.plot(x, y, color='blue', label='y = x^2')  # 绘制折线
plt.xlabel('x')                  # x 轴标签
plt.ylabel('y')                  # y 轴标签
plt.title('Quadratic Function')  # 图表标题
plt.legend()                     # 显示图例
plt.grid(True)                   # 显示网格
plt.show()                       # 显示图像
```

#### 25.画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8,5))                   # 设置图像大小
plt.plot(x, y_sin, color='blue', label='sin(x)')  # sin 曲线
plt.plot(x, y_cos, color='red', label='cos(x)')   # cos 曲线
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Functions')
plt.legend()                                # 显示图例
plt.grid(True)                              # 显示网格
plt.show()                                  # 显示图像
```
